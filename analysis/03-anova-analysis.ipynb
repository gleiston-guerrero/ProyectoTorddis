# 03 --- One-way ANOVA on Statement (5)

Reproduces **Table 12** of the manuscript. For every candidate variable, a one-way ANOVA is applied grouping observations by the value of Statement (5).

**Input:** `../data/raw/likert-effectiveness.csv`
**Output:** `../data/processed/anova-results.csv`

```python
import numpy as np
import pandas as pd
from scipy import stats

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)

df = pd.read_csv('../data/raw/likert-effectiveness.csv', comment='#')

target = 'statement_5'
candidates = ['tutor_age', 'supervision_time', 'student_age',
              'statement_1', 'statement_2', 'statement_3', 'statement_4',
              'statement_6', 'statement_7']

rows = []
for var in candidates:
    groups = [g[var].dropna().to_numpy() for _, g in df.groupby(target)]
    groups = [g for g in groups if len(g) > 1]
    if len(groups) < 2:
        continue
    f, p = stats.f_oneway(*groups)
    if p < 0.01:
        conclusion = 'highly significant'
    elif p < 0.05:
        conclusion = 'significant'
    else:
        conclusion = 'not significant'
    rows.append({'variable': var,
                 'F_statistic': round(f, 4),
                 'p_value': round(p, 4),
                 'conclusion': conclusion})

result = pd.DataFrame(rows)
result.to_csv('../data/processed/anova-results.csv', index=False)
result
```
